```python
#number of Mobiles in each Brand according to Selling Price in ascending order
Mobiles= ak.groupby('Brand').size()
data = (ak.groupby('Brand')['Selling Price'].sum())/Mobiles
data.sort_values(ascending = False)
```

```python
PriceUnder20000 = ak[ak['Original Price']<20000]
PriceUnder20000['Brand'].value_counts().plot(kind = 'bar', color=sns.color_palette('viridis'))
plt.show()
```

```python
#Average of original price in each brand
plt.figure(figsize=(12,6))
sns.boxplot(x = 'Brand', y='Original Price', data= ak)
plt.xticks(rotation = 90)
plt.show()
```

```python
## Total Rating from each Brand
RatingsByBrand = pd.DataFrame(ak.groupby('Brand', as_index = False)['Rating'].sum())
sns.set_theme(style = 'whitegrid',color_codes= True)
a = sns.barplot(x='Brand', y='Rating', data=RatingsByBrand.sort_values(by='Rating'))
a.set_xlabel('Brand', fontsize=30)
a.set_ylabel('Rating',fontsize=20)
plt.xticks(rotation=75)
plt.rcParams['figure.figsize']=(12,8)
plt.title('Plot of Brand vs Total Rating', fontsize = 15)
plt.show()
```

```python
Top10Rating= ak[['Brand', 'Model', 'Rating', 'Selling Price']].nlargest(n=15, columns = 'Rating')
Top10Rating
```

```python
#selling price
ToatalSellingPrice= ak['Selling Price'].sum()
ToatalSellingPrice
```

```python
#original price
TotalOriginalPrice=ak['Original Price'].sum()
TotalOriginalPrice
```

```python
##Piechart for prices selling price vs original price
plt.figure(figsize=(13,6))
a= TotalOriginalPrice - ToatalSellingPrice
data = [a,ToatalSellingPrice]
labels = ['original price', 'selling price']
color = sns.color_palette('Set1')
plt.pie(data, labels = labels, colors = color, autopct = '%.0f%%',shadow=True)
plt.show()
```

```python
#most mobiles form which brand
mobile= ak.groupby('Brand').size().sort_values(ascending =False)
mobile.head(8).plot(kind = 'bar', figsize=(10,6), color = sns.color_palette('colorblind'))
plt.show()
```

```python
#count memory 
sns.countplot(x='Memory',data=ak)
plt.show()
```

```python
#Brand and Rating relate to each other.
sns.relplot(x='Rating',y='Brand',data=ak)
plt.show()
```

```python
#Total Original Price from each Brand
OriginalPricebyBrand = pd.DataFrame(ak.groupby('Brand', as_index= False)['Original Price'].sum())
sns.set_theme(style = 'whitegrid',color_codes= True)
v= sns.barplot(x='Brand', y='Original Price', data=OriginalPricebyBrand.sort_values(by='Original Price'))
v.set_xlabel('Brand',fontsize=30)
v.set_ylabel('Original Price',fontsize=20)
plt.xticks(rotation=65)
plt.rcParams['figure.figsize']=(12,8)
plt.title('Plot of Brand vs Total Original Price ', fontsize = 20)
plt.show()
```
